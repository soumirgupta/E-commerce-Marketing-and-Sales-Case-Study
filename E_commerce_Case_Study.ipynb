<h2><b>3. Identify periods with the strongest and weakest retention rates. What strategies could be implemented to improve retention during weaker months?

```python
#Creating a set with month name and unique sales data with month wise
monthly_orders={}
for month in sales['Month'].unique():
  monthly_orders[month] = (sales[sales['Month'] == month].drop_duplicates(subset=['CustomerID']))
```

```python
#Calculating retention rate from comparing m with m+1 month
retension_rate={}
rows=[]
for month in range(len(months)-1):
  x_month=months[month]
  x_plus_month=months[month+1]
  current_retention_rate=pd.merge(monthly_orders[x_month],monthly_orders[x_plus_month],on='CustomerID',how='inner')
  retension_rate[x_month]=round(len(current_retention_rate)/len(monthly_orders[x_month]),2)
  rows.append({
        'month': x_month,
        'retention_rate': round(len(current_retention_rate)/len(monthly_orders[x_month]),2)
    })
retention_df = pd.DataFrame(rows)
```

```python
retention_df
```

```python
#Graph for Monthly Customer Retention rate
sns.barplot(data=retention_df,x='month',y='retention_rate')
plt.xticks(rotation=45)
plt.title('Monthly Customer Retention rate')
plt.xlabel('Months')
plt.ylabel("Customer Retention rate ")
```

```python
Low_ret_rate=retention_df['retention_rate'].min()
High_ret_rate=retention_df['retention_rate'].max()
Low_ret_rate_month=retention_df.loc[retention_df['retention_rate'].idxmin(),'month']
High_ret_rate_month=retention_df.loc[retention_df['retention_rate'].idxmax(),'month']
print(f"The lowest retention rate month is: {Low_ret_rate_month} and rate is: {Low_ret_rate}")
print(f"The highest retention rate month is: {High_ret_rate_month} and rate is: {High_ret_rate}")
```

<h2><b>8. Analyze the relationship between monthly marketing spend and revenue. Are there any months where marketing efforts yielded disproportionately high or low returns? How can marketing strategies be adjusted to improve ROI?

```python
#Changing the data type of Date column from maeketing to date type
marketing['Date']=pd.to_datetime(marketing['Date'])
#Generating a column with short month name
marketing['Month']=marketing['Date'].dt.strftime('%b')
marketing.head()
```

```python
#Grouping the data on month and taking the total month marketing spends
marketing_summary=marketing.groupby('Month').agg(Offline_Spend=('Offline_Spend','sum'),Online_Spend=('Online_Spend','sum')).reset_index()
marketing_summary.head(12)
```

```python
#Grouping by month in sales table and taking sum of revenue
sales_summary=sales.groupby('Month').agg(MonthlyRevenue=('Total_Order_Cost','sum')).reset_index()
sales_summary
```

```python
#Merging the data marketing with sales to prepare base for ROI calclation
marketing_revenue_summary=marketing_summary.merge(sales_summary,on='Month',how='left')
marketing_revenue_summary['Total_Marketing_Spend']=marketing_revenue_summary['Offline_Spend']+marketing_revenue_summary['Online_Spend']
marketing_revenue_summary
```

<h2><b>10. Segment customers into groups such as Premium, Gold, Silver, and Standard. What targeted strategies can be developed for each segment to improve retention and revenue? (Use RFM segmentation techniques)

```python
sales.head()
```

```python
# Hint: This should be one day after the max Transaction_Date in your data
today = sales['Transaction_Date'].max()+pd.Timedelta(days=1)  # Example placeholder

# STEP 2: Group data by CustomerID and calculate:

# - Recency: days since last purchase
# - Frequency: number of purchases
# - Monetary: total value spent

sales_rfm = sales.groupby('CustomerID').agg({
    'Transaction_Date': lambda x: (today - x.max()).days,   # Recency
    'Transaction_ID': 'count',                              # Frequency
    'Total_Order_Cost': 'sum'                                        # Monetary
}).reset_index()
```

```python
# Renaming the columns for clarity
sales_rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
sales_rfm.head()
```

```python
# Define example thresholds (can be customized or made dynamic)
r_bins = [0, 100, 200, 300, np.inf]
f_bins = [0, 40, 70, 100, np.inf]
m_bins = [0, 4000, 7000, 10000, np.inf]

# Assign segment scores using pd.cut (you can replace with qcut for quantile-based binning)

sales_rfm['R_Score'] = pd.cut(sales_rfm['Recency'], bins=r_bins, labels=False) + 1
sales_rfm['F_Score'] = pd.cut(sales_rfm['Frequency'], bins=f_bins, labels=False) + 1
sales_rfm['M_Score'] = pd.cut(sales_rfm['Monetary'], bins=m_bins, labels=False) + 1
```

```python
sales_rfm.head()
```

```python
# STEP 4: Combine R, F, M scores to compute an overall segment

# Hint: You can use simple addition, weighted sum
sales_rfm['RFM_Score'] = sales_rfm['R_Score'] + sales_rfm['F_Score'] + sales_rfm['M_Score']

sales_rfm.head()
```

```python
# Create a new customer segment based on RFM_Score thresholds
def classify_customer(score):
    if score <= 6:
        return 'Standard'
    elif score <= 9:
        return 'Silver'
    elif score <= 11:
        return 'Gold'
    else:
        return 'Premium'
```

```python
#Generating Customer Segment column by applying a method
sales_rfm['Customer_Segment'] = sales_rfm['RFM_Score'].apply(classify_customer)
sales_rfm.head()
```

```python
#Grouping customer segment wise count
customer_segment_count=sales_rfm['Customer_Segment'].value_counts().reset_index()
customer_segment_count
```
